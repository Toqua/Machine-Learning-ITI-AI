```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = 2 * math.pi * np.random.rand(25,1)
x = np.sort(x, axis =0 )
y = np.sin(x).ravel()
t = y + 0.01-np.random.rand(25)
```

```python
plt.plot(x,t,'*')
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(3)
xlin = poly.fit_transform(x)

from sklearn.linear_model import LinearRegression
m = LinearRegression()
m = m.fit(xlin,t)
y_new = m.predict(xlin)
```

```python
plt.plot(x,y,'g-')
plt.plot(x,t,'b*')
plt.plot(x,y_new,'r-')
```
